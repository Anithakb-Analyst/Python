## Exploratory Data Analysis (EDA) and Visualization

### Campaign Performance:

```python
# Scatter plot to visualize the relationship between Acquisition_Cost and ROI
```

```python
from matplotlib.typing import ColourType
plt.figure(figsize=(7, 6))
plt.scatter(df['Acquisition_Cost'], df['ROI'],edgecolors='black')
plt.title('Acquisition Cost vs. ROI')
plt.xlabel('Acquisition_Cost')
plt.ylabel('ROI')
plt.tight_layout()
plt.show()
```

```python
# Bar chart to visualize the average Conversion_Rate for different Channel_Used, categorized by Campaign_Type.
```

```python
plt.figure(figsize=(8, 6))

df.groupby(['Channel_Used', 'Campaign_Type'])['Conversion_Rate'] \
  .mean() \
  .unstack() \
  .plot(kind='bar', stacked=True)

plt.xlabel('Channel Used')
plt.ylabel('Average Conversion Rate')
plt.title('Average Conversion Rate by Channel and Campaign Type')
plt.xticks(rotation=45)
plt.legend(title='Campaign Type')
plt.tight_layout()
plt.show()
```

```python
# Distribution of Engagement_Score across different Campaign_Type using a box plot:
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Campaign_Type', y='Engagement_Score', data=df, color= 'green', width = 0.25)
plt.title('Engagement Score Distribution by Campaign Type')
plt.xlabel('Campaign Type')
plt.ylabel('Engagement Score')
plt.tight_layout
plt.show()
```

```python
# ROI by Company using a bar chart:
```

```python
avg_roi = df.groupby('Company')['ROI'].mean()
plt.figure(figsize=(7, 6))
avg_roi.plot(kind='bar', color='brown', width=0.3)
plt.xlabel('Company')
plt.ylabel('Average ROI')
plt.title('Average ROI by Company')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Correlation between Engagement_Score and Conversion_Rate using a heatmap:
```

```python
Correlation= df[['Engagement_Score', 'Conversion_Rate']].corr()
plt.figure(figsize=(7, 6))
sns.heatmap(Correlation, annot=True, cmap='coolwarm', linewidths=0.5,fmt='.2f',)
plt.xlabel('Engagement_Score')
plt.ylabel('Conversion_Rate')
plt.title('Correlation between Engagement_Score and Conversion_Rate')
plt.tight_layout()
plt.show()
```

## Customer Segmentation:

```python
# Distribution of Target_Audience:
```

```python
plt.figure(figsize=(7, 6))
df.Target_Audience.value_counts().plot(kind='bar')
plt.title('Distribution of Target_Audience')
plt.xlabel('Target_Audience')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Identification of Customer_Segment has the highest Conversion_Rate for each Language using a bar chart:
```

```python
plt.figure(figsize=(8, 6))
df.groupby(['Language', 'Customer_Segment'])['Conversion_Rate'].mean().unstack() .plot(kind='bar', stacked=True)
plt.xlabel('Language')
plt.ylabel('Average Conversion Rate')
plt.title('Average Conversion Rate by Language and Customer Segment')
plt.tight_layout()
plt.show()
```

```python
# Acquisition_Cost across each Customer_Segment,categorized by Channel_Used, using a box plot:
```

```python
plt.figure(figsize=(7, 6))
sns.boxplot(x='Customer_Segment', y='Acquisition_Cost', data=df, hue='Channel_Used', palette='Set2')
plt.xlabel('Customer_Segment')
plt.ylabel('Acquisition_Cost')
plt.legend(title='Channel_Used', loc='center right')
plt.title('Acquisition Cost by Customer Segment and Channel Used')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# average Conversion_Rate by Language using a bar chart:
```

```python
plt.figure(figsize=(7, 6))
df.groupby('Language')['Conversion_Rate'].mean().plot(kind='bar', color='pink')
plt.xlabel('Language')
plt.ylabel('Average Conversion Rate')
plt.title('Average Conversion Rate by Language')
plt.tight_layout()
plt.show()
```

## Channel Effectiveness:

```python
# Comparison of Engagement_Score for different Channels_Used, segmented by Campaign_Type, using a bar chart:
```

```python
plt.figure(figsize=(7, 6))
df.groupby(['Channel_Used', 'Campaign_Type'])['Engagement_Score'].mean().unstack().plot(kind='bar',stacked = True)
plt.title('Engagement Score Comparison by Channel and Campaign Type')
plt.xlabel('Channel_Used')
plt.ylabel('Average Engagement Score')
plt.tight_layout()
plt.show()
```

```python
# Distribution of total ROI across different Channels_Used using a pie chart:
```

```python
plt.figure(figsize=(7, 6))
df.groupby('Channel_Used')['ROI'].sum().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Total ROI by Channel')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Relationship between Clicks and Impressions for each Channel_Used :
```

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(data=df,x='Impressions',y='Clicks',hue='Channel_Used',style='Channel_Used',s=80,alpha=0.7)
plt.xlabel('Impressions')
plt.ylabel('Clicks')
plt.title('Clicks vs Impressions by Channel')
plt.tight_layout()
plt.show()

```

## Time-Based Analysis:

```python
# Distribution of Duration using Histogram:
```

```python
plt.figure(figsize=(7, 6))
sns.histplot(df['Duration'], bins=10, kde=True, color='orange')
plt.xlabel('Duration')
plt.ylabel('Frequency',)
plt.title('Distribution of Duration')
plt.tight_layout()
plt.show()
```

```python
# Analysis of Overall Conversion_Rate over Date for each Company using a line chart:
```

```python
plt.figure(figsize=(7, 6))
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
avg_conv = df.groupby(['Date', 'Company'], as_index=False)['Conversion_Rate'].mean()

sns.lineplot(
    data=avg_conv,
    x='Date',
    y='Conversion_Rate',
    hue='Company'
)
plt.xlabel('Date')
plt.ylabel('Conversion_Rate')
plt.xticks(rotation=45)
plt.title('Overall Conversion_Rate over Date for each Company')
plt.tight_layout()
plt.show()
```

```python
## Trend of Engagement_Score over Date with a line chart:
```

```python
from numpy._core.defchararray import title
plt.figure(figsize=(7, 6))
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
avg_engagement = df.groupby(['Date'], as_index=False)['Engagement_Score'].mean()
sns.lineplot(data=avg_engagement,x='Date',y='Engagement_Score')
plt.xlabel('Date')
plt.ylabel('Engagement_Score')
plt.title('Trend of Engagement_Score over Date')
plt.tight_layout()
plt.show()
```

```python
# Proportion of ROI by Location using a pie chart :
```
